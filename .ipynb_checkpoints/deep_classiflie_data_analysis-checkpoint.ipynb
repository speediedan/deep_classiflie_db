```python
#manually analyzed the relevant class distributions to determine the appropriate bounds. Not worth the effort to automate at this point
#initial falsehood distribution
wp_sql= """select (length(statement_text)-length(replace(statement_text,' ',''))+1) as wc from falsehoods"""
#initial truths distribution
fb_sql = """select (length(statement_text)-length(replace(statement_text,' ',''))+1) as wc from truths"""
#truncated falsehood distribution 
t_wp_sql= """select (length(statement_text)-length(replace(statement_text,' ',''))+1) as wc from pt_converged_dt_falsehoods"""
#transformed truths distribution
t_fb_sql = """select (length(statement_text)-length(replace(statement_text,' ',''))+1) as wc from pt_converged_dt_truths"""
f, (ax1, ax2) = plt.subplots(nrows=1, ncols=2,figsize=(12, 5), sharex=True, sharey=True)
f.suptitle('Distribution transformation', y=1.05,fontsize=20)
ax1.set_title('Before distribution convergence', fontsize=18)
ax2.set_title('After distribution convergence', fontsize=18)
ax1.set_xlabel('word count', fontsize=14)
ax2.set_xlabel('word count', fontsize=14)
ax1.set_ylabel('density', fontsize=14)
sql_engine = sqla.create_engine(connect_string)
wp_df, fb_df, t_wp_df, t_fb_df = pd.DataFrame(), pd.DataFrame(), pd.DataFrame(), pd.DataFrame()
wp_df = pd.read_sql_query(wp_sql, sql_engine)
wp_df = wp_df.rename(columns={"wc": "falsehoods"})
fb_df = pd.read_sql_query(fb_sql, sql_engine)
fb_df = fb_df.rename(columns={"wc": "truths"})
t_wp_df = pd.read_sql_query(t_wp_sql, sql_engine)
t_wp_df = t_wp_df.rename(columns={"wc": "falsehoods"})
t_fb_df = pd.read_sql_query(t_fb_sql, sql_engine)
t_fb_df = t_fb_df.rename(columns={"wc": "truths"})
for class_df in [wp_df['falsehoods'], fb_df['truths']]:
   sns.kdeplot(class_df, shade=True, ax=ax1)
for class_df in [t_wp_df['falsehoods'], t_fb_df['truths']]:
   sns.kdeplot(class_df, shade=True, ax=ax2)
sql_engine.dispose()
```

```python
def load_json(filename):
    # open the file as read only
    with open(filename, 'r') as f:
        # read all text
        json_data = json.load(f)
    return json_data

#ad hoc analysis/validation of train, val, test distributions for each label
def distribution_analysis():
    datasets = ['train', 'val', 'test']
    classes = ['truths', 'falsehoods']
    sql = {}
    ds_metadict = load_json(ds_metadata)
    sql['falsehoods_train_sql'] = f"""with falsehood_train_ds AS 
    (select (length(statement_text)-length(replace(statement_text,' ',''))+1) as wc from pt_converged_dt_falsehoods where s_date between STR_TO_DATE('{ds_metadict['train_start_date']}','%Y-%m-%d') and STR_TO_DATE('{ds_metadict['train_end_date']}','%Y-%m-%d')) 
    select * from falsehood_train_ds"""
    sql['falsehoods_val_sql'] = f"""with falsehood_val_ds AS 
    (select (length(statement_text)-length(replace(statement_text,' ',''))+1) as wc from pt_converged_dt_falsehoods where s_date between STR_TO_DATE('{ds_metadict['val_start_date']}','%Y-%m-%d') and STR_TO_DATE('{ds_metadict['val_end_date']}','%Y-%m-%d')) 
    select * from falsehood_val_ds"""
    sql['falsehoods_test_sql'] = f"""with falsehood_test_ds AS 
    (select (length(statement_text)-length(replace(statement_text,' ',''))+1) as wc from pt_converged_dt_falsehoods where s_date between STR_TO_DATE('{ds_metadict['test_start_date']}','%Y-%m-%d') and STR_TO_DATE('{ds_metadict['test_end_date']}','%Y-%m-%d')) 
    select * from falsehood_test_ds"""
    sql['truths_train_sql'] = f"""with truths_train_ds AS
    (select (length(statement_text)-length(replace(statement_text,' ',''))+1) as wc from pt_converged_dt_truths where s_date between STR_TO_DATE('{ds_metadict['train_start_date']}','%Y-%m-%d') and STR_TO_DATE('{ds_metadict['train_end_date']}','%Y-%m-%d'))
    select * from truths_train_ds"""
    sql['truths_val_sql'] = f"""with truths_val_ds AS
    (select (length(statement_text)-length(replace(statement_text,' ',''))+1) as wc from pt_converged_dt_truths where s_date between STR_TO_DATE('{ds_metadict['val_start_date']}','%Y-%m-%d') and STR_TO_DATE('{ds_metadict['val_end_date']}','%Y-%m-%d'))
    select * from truths_val_ds"""
    sql['truths_test_sql'] = f"""with truths_test_ds AS
    (select (length(statement_text)-length(replace(statement_text,' ',''))+1) as wc from pt_converged_dt_truths where s_date between STR_TO_DATE('{ds_metadict['test_start_date']}','%Y-%m-%d') and STR_TO_DATE('{ds_metadict['test_end_date']}','%Y-%m-%d'))
    select * from truths_test_ds"""
    sql_engine = sqla.create_engine(connect_string)
    ds_dict = {}
    for ds in datasets:
        ds_dict[ds] = {}
        ds_dict[ds]['dfs'] = {}
        for c in classes:
            dfs = {}
            class_ds_id = f"{c}_{ds}"
            df_name = f"{class_ds_id}_df"
            sql_name = f"{class_ds_id}_sql"
            dfs[df_name] = pd.read_sql_query(sql[sql_name], sql_engine)
            dfs[df_name] = dfs[df_name].rename(columns={"wc": f"{class_ds_id}"})
            ds_id = (dfs[df_name])[class_ds_id]
            ds_dict[ds]['dfs'][c] = ds_id
            ds_id = ds_id.astype('double')
        sql_engine.dispose()
    return ds_dict
    
def print_ds_dict_summary(ds_dict):
    for ds, subd in ds_dict.items():
        print(f"{ds} dataset summary:")
        print(f"dataset {ds} kde diff determined to be {subd['kde_diff']}")
        for cls, df in subd['dfs'].items():
            print(f"{cls} dataframe:")
            print(df.describe())
        
def plot_dists(ds_dict: Dict) -> None:
    axes = []
    f, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 5), sharex=True, sharey=True)
    datasets = [ds for ds in ds_dict.keys()]
    suptitle = f'{"/".join(datasets)} Distributions'
    f.suptitle(suptitle, y=1.05, fontsize=20)
    axes_idx = 0
    axes[0].set_ylabel('density', fontsize=14)
    for ds, subd in ds_dict.items():
        axes[axes_idx].set_title(ds, fontsize=18)
        axes[axes_idx].set_xlabel('word count', fontsize=14)
        for cls, df in subd['dfs'].items():
            sns.kdeplot(df, shade=True, ax=axes[axes_idx])
        axes_idx+=1
```

```python
# N.B. the truths dataset distribution is temporally non-stationary (seen wrt word count here) with divergence due in great part to twitter dataset contribution
# To improve desired learning signal (reduce wc-based artifacts), the dataprep pipeline could be altered to use separate training/val/test target_dist views for maximal dist match wrt word count
ds_dict = distribution_analysis()
plot_dists(ds_dict)
```

```python
sql_engine = sqla.create_engine(connect_string)
def gen_rpt(sql, rpt_type):
    model_analysis_rpt_df  = pd.read_sql_query(sql, sql_engine)
    model_analysis_rpt_df = model_analysis_rpt_df.rename(columns={"yweeks": "week"})
    model_analysis_rpt_df = model_analysis_rpt_df.rename(columns={"tp_ratio": "true positives"})
    model_analysis_rpt_df = model_analysis_rpt_df.rename(columns={"tn_ratio": "true negatives"})
    model_analysis_rpt_df = model_analysis_rpt_df.rename(columns={"fp_ratio": "false positives"})
    model_analysis_rpt_df = model_analysis_rpt_df.rename(columns={"fn_ratio": "false negatives"})
    labels = list(model_analysis_rpt_df.columns)[1:]
    # print(model_analysis_rpt_df.describe())
    fig, ax = plt.subplots(figsize=(12, 5))
    pal = ["#258f5c", "#12e07d", "#e01212", "#ffb8b8"]
    ax.stackplot(model_analysis_rpt_df['week'], model_analysis_rpt_df['true positives'],model_analysis_rpt_df['true negatives'],
                 model_analysis_rpt_df['false positives'],model_analysis_rpt_df['false negatives'],labels=labels, colors=pal, alpha=0.4)
    plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
    ax.set_ylabel('percent of examples', fontsize=14)
    ax.set_title(f'Confusion Matrix For Falsehood Classification ({rpt_type})', fontsize=18)
    ax.set_xlabel('week', fontsize=14)
    plt.margins(x=0, y=0)
    plt.xticks(rotation=90)
    plt.show()
tgt_reports = constants.LATEST_MODEL_REPORTS
tgt_reports.extend(constants.TEST_CMATRICES)
for t in tgt_reports:
    sql = f"select * from {t}"
    gen_rpt(sql,t)
```

```python
def gen_rpt(sql, rpt_type):
    model_analysis_rpt_df  = pd.read_sql_query(sql, sql_engine)
    model_analysis_rpt_df = model_analysis_rpt_df.rename(columns={"confidence_bucket": "confidence bucket"})
    model_analysis_rpt_df = model_analysis_rpt_df.rename(columns={"tp_ratio": "true positives"})
    model_analysis_rpt_df = model_analysis_rpt_df.rename(columns={"tn_ratio": "true negatives"})
    model_analysis_rpt_df = model_analysis_rpt_df.rename(columns={"fp_ratio": "false positives"})
    model_analysis_rpt_df = model_analysis_rpt_df.rename(columns={"fn_ratio": "false negatives"})
    labels = list(model_analysis_rpt_df.columns)[1:]
    #print(model_analysis_rpt_df.describe())
    fig, ax = plt.subplots(figsize=(12, 5))
    pal = ["#258f5c", "#12e07d", "#e01212", "#ffb8b8"]
    ax.stackplot(model_analysis_rpt_df['confidence bucket'], model_analysis_rpt_df['true positives'],model_analysis_rpt_df['true negatives'],
                 model_analysis_rpt_df['false positives'],model_analysis_rpt_df['false negatives'],labels=labels, colors=pal, alpha=0.4)
    plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
    ax.set_ylabel('percent of examples', fontsize=14)
    ax.set_title(f'Confusion Matrix For Falsehood Classification ({rpt_type})', fontsize=18)
    ax.set_xlabel('confidence bucket', fontsize=14)
    plt.xticks(rotation=90)
    plt.margins(x=0, y=0)
    plt.show()
for t in constants.TEST_CONF_CMATRICES:
    sql = f"select confidence_bucket, tp_ratio, tn_ratio, fp_ratio, fn_ratio from {t}"
    gen_rpt(sql,t)
```
